## Question One: First Duplicate in List
**Description**  
Given a list of integers, return the **first value that appears more than once**. If there are multiple duplicates, return the one that appears **first** in the list. If no duplicate exists, return `-1`.

**Examples**
```python
Input: nums = [3, 1, 4, 2, 5, 1, 6]
Output: 1
```
```python
Input: nums = [7, 8, 9, 10]
Output: -1
```
```python
Input: nums = [4, 5, 6, 4, 6]
Output: 4
```

**Question 1 Starter Code**

````python
from typing import List

def first_duplicate(nums: List[int]) -> int:
    seen = set()
    
    for num in nums:
        if num in seen:
            return num
        seen.add(num)
    
    return -1
````

````python
nums_1 = [3, 1, 4, 2, 5, 1, 6]
nums_2 = [7, 8, 9, 10]
nums_3 = [4, 5, 6, 4, 6]

print(first_duplicate(nums_1))
print(first_duplicate(nums_2))
print(first_duplicate(nums_3))
````


-   Code the solution to your assigned problem in Python (code chunk). Note: each problem can be solved more simply if you use an abstract data type that is suitable for that problem. Using that try to find the best time and space complexity solution!


````python
from typing import List

def first_duplicate(nums: List[int]) -> int:
    seen = set()
    for x in nums:
        if x in seen:
            return x
        seen.add(x)
    return -1
````


-   Explain why your solution works


-  The algorithm uses a set containing all of the values encountered so far while scanning the list from left to right. When a new value is read, it checks whether it is already in the set. If it is, then the value must have appeared earlier, so it is tagged as a duplicate. Because the values are scanned in sequential order, the first time we detect a repeated value is exactly the duplicate that “appears first” according to the problem statement. If no duplicates are found by the end, we return -1.




-   Explain the problem’s time and space complexity


- Time complexity: O(n) where n is the length of the list, because each element is processed once and set membership checks are O(1) on average. 
- Space complexity: O(n) in the worst case, because the seen set may store every element if there are no duplicates.


-   Explain the thinking to an alternative solution (no coding required, but a classmate reading this should be able to code it up based off your text)


A simple alternative solution would be a nested loop approach:

- For each index i from 0 to n-1, compare nums[i] with every later element nums[j] for j > i.

- If you find any match, that means nums[i] is a duplicate.

- Return the first nums[i] for which a match exists (because i increases left to right, this ensures you return the one that appears first in the list).

- If you finish all comparisons without finding a match, return -1.

- This alternative uses O(1) extra space, but costs O(n^2) time.
