#### Starter Code for Question 1

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def is_duplicate(root: TreeNode) -> int:
    if not root:
        return -1
    
    seen = set()
    queue = [root]
    
    while queue:
        node = queue.pop(0)
        if node.val in seen:
            return node.val
        seen.add(node.val)
        
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    
    return -1
```

```python
#Function to build binary tree from inputs
def build_tree(values):
    if not values:
        return None
    root = TreeNode(values[0])
    queue = [root]
    i = 1
    while queue and i < len(values):
        node = queue.pop(0)
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    return root

#Examples
print("Examples:\n")

#Example #1
root = build_tree([1, 2, 2, 3, 5, 6, 7])
print("Example 1 \nInput", [1, 2, 2, 3, 5, 6, 7], "\nOutput:", is_duplicate(root),"\n")

#Example #2
root = build_tree([1, 10, 2, 3, 10, 12, 12])
print("Example 2 \nInput", [1, 10, 2, 3, 10, 12, 12], "\nOutput:", is_duplicate(root),"\n")

#Example #3
root = build_tree([10, 9, 8, 7])
print("Example 3 \nInput", [10, 9, 8, 7], "\nOutput:", is_duplicate(root))
```

```python
# The problem asks us to check whether a string of brackets (e.g., {} [] ()) is properly balanced, meaning every opening bracket has a corresponding closing bracket.
# Every opening bracket must have a closing bracket of the same type, and the brackets must also close in the correct nested order to be considered valid.
```

```python
# New Example: )[]{})   will return "False" because the first character is a closing bracket without an opening bracket.
# Partner's Example 2: {()[]   will return "False" because the first character is an opening bracket without a closing bracket.
```


-   Copy the solution your partner wrote. 


```python
def is_valid_brackets(s: str) -> bool:
    open_brackets = []
    for char in s:
        if char in ['(', '[', '{']:
            open_brackets.append(char)
        else:
            if (open_brackets[-1] + char) in ['[]','()', '{}']:
                open_brackets.pop()
            else:
                return False
    # Check if open_brackets is empty
    if not open_brackets:
        return True
    else:
        return False
    pass

print(is_valid_brackets('([]{})'))          # True
print(is_valid_brackets('([)]'))            # False
print(is_valid_brackets('()[]{}'))          # True
print(is_valid_brackets('[{]}'))            # False
print(is_valid_brackets('{[()()]}'))        # True
print(is_valid_brackets('{()[]'))           # False
print(is_valid_brackets('('))               # False   
print(is_valid_brackets('(()'))             # False  
```


-   Explain why their solution works in your own words.


```python
# The solution works because it uses a stack to keep track of each opening bracket.
# Each time a closing bracket appears, it checks to see if it matches the most recent opening bracket.
# If all brackets match correctly and the stack is empty at the end (i.e., there are no open brackets remaining), then it considers the sequence as valid.
```


-   Explain the problem’s time and space complexity in your own words.


```python
#Time Complexity: O(n) — each character in the string is processed once.
#Space Complexity: O(n) — in the worst case, all opening brackets are stored in the stack before being closed.
```
